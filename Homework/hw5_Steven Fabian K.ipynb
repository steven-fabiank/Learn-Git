```python
import pandas as pd

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
# membuang kolom Passenger ID, Name, Ticket number, Fare, Cabin, dan Embarking ports

df_train.drop(['PassengerId','Name','Ticket','Fare','Cabin','Embarked'],axis=1,inplace=True)
df_train.head()
```

```python
df_test.drop(['PassengerId','Name','Ticket','Fare','Cabin','Embarked'],axis=1,inplace=True)
df_test.head()
```

```python
df_train.info()
```

```python
# membuang baris tanpa nilai umur (age NA)

df_train.dropna(inplace=True)
df_train.info()
```

```python
df_test.dropna(inplace=True)
df_test.info()
```

```python
# membinerkan jenis kelamin

df_train = pd.get_dummies(df_train)
df_train.head()
```

```python
df_test = pd.get_dummies(df_test)
df_test.head()
```

```python
#scaling dataset Train

x = df_train[['Pclass','Age','SibSp','Parch','Sex_female','Sex_male']]
y = df_train['Survived']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data,columns=x.columns)
scaled_data
```

```python
# menggunakan dataset Train untuk training dan testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_data,y,test_size = 0.9,random_state = 1)

from sklearn import linear_model
model = linear_model.LogisticRegression()
model.fit(x_train,y_train)

from sklearn.preprocessing import StandardScaler
model.score(x_test,y_test)
```

```python
# scaling dataset Test dan hasil testingnya

x_test = df_test[['Pclass','Age','SibSp','Parch','Sex_female','Sex_male']]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_test)
scaled_data = scaler.transform(x_test)
scaled_data = pd.DataFrame(scaled_data,columns=x_test.columns)
scaled_data

model.predict(scaled_data)
```
